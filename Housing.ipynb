```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python

df = pd.read_csv("boston_housing.csv", index_col=False)
print(df.head(11))

```

```python
frame2 = df.drop(["CHAS", "NOX", "B", "LSTAT"], axis=1)
print(frame2.tail)
```

Plot all variables compared to AGE as the x-axis

```python
for i in frame2:
    y = frame2[str(i)]
    x = frame2["AGE"]
    plt.hist(frame2, edgecolor="black")
    plt.title("Comparison Age to " + str(i))
    plt.xlabel("AGE")
    plt.ylabel(str(i))   
    plt.show()
```

```python
y = frame2["CRIM"]
x = frame2["PRICE"]
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Price")
plt.ylabel("Crime Rate")
    
m, b = np.polyfit(x, y, 1)

# Add line of best fit to plot
plt.plot(x, m*x + b)
plt.show()
```

```python


y = frame2["CRIM"]
x = frame2["PRICE"]
plt.yscale('log')
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Price")
plt.ylabel("Crime Rate (log)")
    
m, b = np.polyfit(x, y, 1)


plt.show()
```
